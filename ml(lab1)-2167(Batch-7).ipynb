```python
from numpy._core.defchararray import count
mean_score=np.mean(exam_score_array)
median_score=np.median(exam_score_array)
std_score=np.std(exam_score_array)
count_score=np.count_nonzero(exam_score_array)
min_score=np.min(exam_score_array)
max_score=np.max(exam_score_array)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(exam_score_array,bins=100)
plt.xlabel('Exam Score')
plt.ylabel('no.of students')
plt.title('Distribution of Exam Scores')
plt.show()
```

```python
score,counts=np.unique(exam_score_array,return_counts=True)
plt.bar(score,counts,width=1)
plt.xlabel('Exam Score')
plt.ylabel('no.of students')
plt.title('Distribution of Exam Scores')
plt.show()
```

```python
scores,counts=np.unique(exam_score_array,return_counts=True)
plt.pie(counts,labels=scores,autopct='%1.1f%%')
plt.title('pie chart of exam score')
plt.show()
```
