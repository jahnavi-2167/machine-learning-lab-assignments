```python
x_min=x.min()
x_max=x.max()
x_norm=(x-x_min)/(x_max-x_min)
x_norm
```

```python
m=np.random.rand()
c=np.random.rand()
print(m,c)
```

```python
def predict(x,m,c):
  y_pred=m*x+c
  return y_pred
```

```python
#make predition using intial values
y_pred=predict(x,m,c)
y_pred
```

```python
#display first 5 predictions
print("first 5 predictions :")
print(y_pred[:5])
```

```python
#mean square error(cost function)
def compute_error(y,y_pred):
  n=len(y)
  cost=(1/(2*n))* np.sum((y_pred - y)) ** 2
  return cost
```

```python
#compute intial cost
cost = compute_error(y,y_pred)
print("intial cost(mse)", cost)
```

```python
#give parameters
m=0.0
c=0.0
#prediction
y_pred=predict(x,m,c)
#compute cost
cost=compute_error(y,y_pred)
print("cost for m=", m,"and c=",c,"is:",cost)
```

```python
learning_rate=0.01
epochs=100
print("learning rate:",learning_rate)
print("no.of epochs:",epochs)
```

```python
#no.of traning explams
n=len(y)
```

```python
cost_history=[]
```

```python
#gradient descent loop
for epoch in range(epochs):
  #step1:compute prediction
  y_pred=predict(x_norm,m,c)

  #step2:compute gradients
  dm=(1/n)*np.sum((y_pred-y)*x_norm)
  dc=(1/n)*np.sum((y_pred-y))

  #step3:update parameters
  m=m-learning_rate*dm
  c=c-learning_rate*dc

  #step4:compute and store cost
  cost=compute_error(y,y_pred)
  cost_history.append(cost)

  #print cost occasinally
  if epoch%100==0:
    print(f"epoch {epoch}: cost:{cost}")
```

```python
#store final tranied parametrs
final_slop=m
final_intercept=c
final_cost=cost
print("cost":cost)
print("final slop:",final_slop)
print("final intercept:",final_intercept)
```

```python
def predict_salary(Years,m,c,x_norm):
  salary_pred=m*x_norm+c
  return salary_pred
```

```python
#example :predict salary for 5 years of experience
years=5
predicted_salary=predict_salary(
    years,
    final_slop,
    final_intercept,
    x_norm
)
print("predicted salary for",years,"years of experience:",predicted_salary[0][0])
```

```python
#predict outputs using trained model
y_pred_final=predict(x_norm,final_slop,final_intercept)

#calculate mean squared error
mse=np.mean((y - y_pred_final)**2)
print("mean squared error:",mse)
```
